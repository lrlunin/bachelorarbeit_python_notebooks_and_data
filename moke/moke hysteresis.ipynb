```python
I, B, HallErr, avgDiff, errDiff, avgSum, errSum, avgR, errR = np.loadtxt("20220309_LL/ds220107_R1_probe_1_0000.dat", delimiter="	",skiprows=299, unpack=True)
```

```python
%matplotlib inline 
#grob
fig, axs = plt.figure(), plt.gca()
#fig.suptitle('MOENCH detector characteristics')
#plt.subplots_adjust(hspace =0.6)
axes = axs
axes.set_xlim((-165, 165))
#axes.set_title('Soft X-Ray Range')
axes.set_xlabel(r'magnetische Feldstärke $B$ in mT')
axes.set_ylabel(r"Magnetisierungsgrad $\chi$ in \%")
#axes.grid(which='major')
#axes.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)
#axes.xaxis.set_major_formatter(OOMFormatter(order = 0, fformat = "%1.0f"))
#axes.yaxis.set_major_formatter(OOMFormatter(order = 0, fformat = "%1.0f"))
axes.xaxis.set_major_locator(MultipleLocator(50))
axes.xaxis.set_minor_locator(AutoMinorLocator(5))
axes.yaxis.set_major_locator(MultipleLocator(25))
axes.yaxis.set_minor_locator(AutoMinorLocator(5))
axes.plot(B, -1*avgR/np.abs(np.max(avgR))*100)
```
